```python
Name = []
Rating = []
Price =[]
OS =[]
Screen_Res =[]
Warranty =[]
```

```python
for data in tv.findAll('div',class_='_3pLy-c row'):
    specification = tv.find('div', attrs={'class':'fMghEO'})
        
    for each in specification:
        col=each.find_all('li', attrs={'class':'rgWa7D'})
        os_ =col[0].text
        hd_ = col[1].text
        if len(col)>2:
            wr_ = col[2].text
        else:
            wr_ = 'None'
    
        OS.append(os_)
        Screen_Res.append(hd_)
        Warranty.append(wr_)
```
